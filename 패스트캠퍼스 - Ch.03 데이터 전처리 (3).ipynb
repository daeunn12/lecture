# 10. apply, map을 활용한 데이터 변환

```python
df=pd.DataFrame({'a':[1, 2, 3, 4, 5]})
df
```

-문제: a가 2보다 작으면 '2미만', 4보다 작으면 '4미만', 4보다 크면 '4이상'이 저장된 b칼럼을 추가하시오

```python
df['b']=0
df
```

```python
a=df[df['a']<2]
a
```

```python
df['b'][a.index]='2미만'
df
```

```python
a=df[(df['a']>=2)&(df['a']<4)]
a
```

```python
df['b'][a.index]='4미만'
df
```

```python
a=df[df['a']>=4]
a
```

```python
df['b'][a.index]='4이상'
df
```

- 함수 + apply를 이용한 해결

```python
def case_function(x):
    if x < 2:
        return '2미만'
    elif x < 4:
        return '4미만'
    else:
        return '4이상'
```

```python
df['c']=df['a'].apply(case_function)
df
```

- 문제: a가 1이면 'one', 2이면 'two', 3이면 'three', 4이면 'four', 5이면 'five'를 출력하는 칼럼 d를 만드시오

```python
df
```

-사용자 정의함수를 이용한 해결 방법

```python
def function(x):
    if x == 1:
        return 'one'
    elif x == 2:
        return 'two'
    elif x == 3:
        return 'three'
    elif x == 4:
        return 'four'
    elif x== 5:
        return 'five'
```

```python
df['d']=df['a'].apply(function)
df
```

- map를 이용한 해결 방법

```python
a = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five'}
```

```python
df['e']=df['a'].map(a)
df
```

- Left join, Inner join 사용하기

-Left join: df1의 ID가 1, 2, 3 df2의 ID가 3,4,5가 있다면 df1의 1, 2, 3과 df2의 3의 정보를 가져오는 것

-Inner join: 3의 df1과 df2의 정보를 가져오는 것

```python
df1=pd.DataFrame({'ID':[1, 2, 3, 4, 5], '성별':['F', 'M', 'F', 'M', 'F'], '나이':[20, 30, 40, 25, 42]})
df2=pd.DataFrame({'ID':[3, 4, 5, 6, 7], '키':[160.5, 170.3, 180.1, 142.3, 153.7], '몸무게':[45.1, 50.3, 72.1, 38, 42]})
```

-문제: 성별과 나이가 확인된 유저들을 대상으로 키와 몸무게의 정보를 결합하시오

```python
pd.merge(df1, df2, how='left', on='ID')
```

-문제: 키와 몸무게가 확인된 유저들을 대상으로 성별과 나이의 정보를 결합하시오

```python
# 방법1
pd.merge(df2, df1, how='left', on='ID')
```

```python
# 방법2
pd.merge(df1, df2, how='right', on='ID')
```

-문제: 키, 몸무게, 성별, 나이 정보가 모두 확인된 유저들의 정보를 출력하시오

```python
pd.merge(df1, df2, how='inner', on='ID')
```

-문제: 모든 유저들의 정보를 출력하시오
