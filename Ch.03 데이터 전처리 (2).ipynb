# 5. loc와 iloc를 이용한 원하는 위치의 데이터 추출

- 데이터 생성

```python
import pandas as pd
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11, 21)], 'c':[i for i in range(21, 31)]})
df
```

- 문제: a,b열을 추출하시오

```python
df['a', 'b']
## a와 b의 각각의 시리즈를 한꺼번에 뽑을 수 없음
```

```python
## 시리즈로 뽑을 수 없어서 데이터프레임 형태로 뽑아야함
df[['a', 'b']]
```

```python
type(df[['a', 'b']])
```

```python
## 인덱스 번호로 뽑을 수 없음
df[0]
```

```python
## loc를 사용해서 출력
df.loc[0]
```

-슬라이스도 가능

```python
df.loc[2:4]
## 리스트의 슬라이스와 다른점: 원하는 인덱스의 값을 다 가져오는 것
```

- 인덱스가 문자로 이루어진 데이터 프레임 생성

```python
index = ['a', 'b', 'd', 'c', 'e', 'f', 'g', 'g', 'h', 'i']
```

```python
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11, 21)], 'c':[i for i in range(21, 31)]}, index=index)
df
```

```python
df.loc[0]
## 0이라는 인덱스가 존재하지 않아 에러가 생김
```

```python
df.loc['g']
```

```python
df.loc['c':]
```

- 문제: 열이 a, c이며 인덱스가 g,i인 데이터를 출력하시오

```python
df
```

```python
df.loc[['g', 'i'], ['a', 'c']]
```

- 문제: 처음부터 5번째 까지의 데이터와 첫번째 열과 세번째 열의 데이터를 추출하시오

```python
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11, 21)], 'c':[i for i in range(21, 31)]})
df               
```

- iloc를 사용한 해결

```python
df.iloc[:5, [0, 2]]
## df.iloc[:5, ['a', 'c']] 에러생김
```

- 처음부터 5번째 까지의 데이터와 첫번째 열과 세번째 열의 데이터를 추출하시오

```python
index = ['a', 'b', 'd', 'c', 'e', 'f', 'g', 'g', 'h', 'i']
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11, 21)], 'c':[i for i in range(21, 31)]}, index=index)
df
```

```python
df.iloc[:5, [0, 2]]
```

# 6. 조건에 맞는 데이터 추출

- 데이터 생성

```python
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11, 21)], 'c':[i for i in range(21, 31)]})
df
```

- 문제: a,c열을 출력하시오

```python
df[['a', 'c']]
```

- 문제: a가 3 이상인 데이터를 출력하시오

```python
df[df['a'] >= 3]
```

- 문제: a가 3 이상인 데이터 중 a,c열만 출력하시오

```python
df[df['a'] >= 3][['a', 'c']]
```

- 문제: a가 3 이상이고, b가 16 미만인 데이터를 출력하시오

```python
df
```

```python
df[(df['a'] >= 3) & (df['b'] < 16)]
```

```python
a = (df['a'] >= 3) & (df['b'] < 16)
a
```

```python
type(a)
```

```python
df[a]
```

- 문제: a가 3 이하이거나 7 이상인 데이터를 출력하시오

```python
df
```

```python
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

- 문제: a가 3 이상이고, b가 16 미만이거나 c가 30인 데이터를 출력하시오

```python
df
```

```python
df[(df['a'] >= 3) & (df['b'] < 16) | (df['c'] == 30)]
```

# 8. 결측값 처리

```python
import numpy as np
df = pd.DataFrame({'a':[1, 1, 3, 4, 5], 'b':[2, 3, np.nan, 3, 4], 'c':[3, 4, 7, 6, 4]})
df
```

- 결측 유무 확인

```python
df.isnull()
```

- 결측값 개수 확인

```python
df.isnull().sum()
```

- 결측값이 포함된 열/행 지우기

-결측값이 포함된 행 지우기

```python
df.dropna()
```

```python
df ## 원래의 df값은 변경되지 않음 -> inplace=True 사용
```

-결측값이 포함된 열 지우기

```python
df.dropna(axis=1)  ## inplace=True 옵션 존재
```

- 결측값을 다른 값으로 대체하기

```python
df
```

-특정값으로 대체하기

```python
df.fillna(0, inplace=True)
df
```

-앞이나 뒤의 숫자로 바꾸기 -> 데이터가 월매출일 경우 앞이나 뒤의 값으로 대체해주면 추세를 따라가는데 보기좋음

```python
df = pd.DataFrame({'a':[1, 1, 3, 4, np.nan], 'b':[2, 3, np.nan, np.nan, 4], 'c':[np.nan, 4, 1, 1, 4]})
df
```

1) 뒤의 값으로 채우기

```python
df.fillna(method='bfill') # inplace=True 옵션 존재
```

2) 앞의 값으로 채우기

```python
df.fillna(method='ffill')
```

-limit 설정

```python
df
```

```python
df.fillna(method='ffill', limit=1)
## limit=1 -> 앞의 값을 한번만 가져오겠다
```

- 문제: 데이터 프레임이 존재하는 결측값들을 뒤의 값으로 대체한 이후 앞의 값으로 대체하시오

```python
df = pd.DataFrame({'a':[1, 1, 3, 4, np.nan], 'b':[2, 3, np.nan, np.nan, 4], 'c':[np.nan, 4, 1, 1, 4]})
df
```

```python
df.fillna(method='bfill', inplace=True)
df
```

```python
df.fillna(method='ffill', inplace=True)
df
```

- 평균값으로 대체

```python
df = pd.DataFrame({'a':[1, 1, 3, 4, np.nan], 'b':[2, 3, np.nan, np.nan, 4], 'c':[np.nan, 4, 1, 1, 4]})
df
```

```python
df.mean()['a']
```

```python
df.mean()
```

- 문제: b,c의 결측값들의 각 열의 평균으로 치환하시오

# 9. 타입 변환

- 데이터 생성

```python
import pandas as pd

df= pd.DataFrame({'판매일':['5/11/21', '5/12/21', '5/13/21', '5/14/21','5/15/21'],
                 '판매량': ['10', '15', '20', '25', '30'], '방문자수':['10', '-', '17', '23', '25'],
                 '기온':['24.1', '24.3', '24.8', '25', '25.4']})
df
```

- 타입 확인

```python
df.dtypes
```

```python
df['판매량 보정']= df['판매량'] + 1 
```

- 문제: 판매량을 정수 형태로 변환하시오

```python
df.astype({'판매량':'int'})
```

```python
df.dtypes ##원래랑 동일함 
```

```python
## 지정해주어야함
df= df.astype({'판매량':'int'})
```

```python
df.dtypes
```

```python
df['판매량 보정']= df['판매량'] + 1 
df
```

- 문제: 방문자수를 숫자 타입으로 변형하시오

```python
## 방문자수의 '-' 때문에 에러가 발생
df.astype({'방문자수':'int'})
```

```python
## 위의 방법 해결하기 위해 pd.to_numeric을 쓰지만 역시 에러가 발생
pd.to_numeric(df['방문자수'])
```

```python
## 위의 에러 해결법
pd.to_numeric(df['방문자수'], errors='coerce')
```

```python
## 이것도 역시 지정해줘야함
df['방문자수']=pd.to_numeric(df['방문자수'], errors='coerce')
```

```python
df.dtypes
```

```python
df
```

```python
## 방문자수를 실수->정수로 변경
df=df.astype({'방문자수':'int'})
```

```python
## 에러가 발생해서 nan 값을 채워줌
df.fillna(0, inplace=True)
```

```python
df=df.astype({'방문자수':'int'})
```

```python
df.dtypes
```

```python
df
```
