### 함수

- 함수에는 기본적으로 우리가 사용했던 내장함수 (built-in function)가 있다
    - len, sum, abs 등
- 이러한 built-in function뿐만 아니라 우리가 직접 function을 define해서 사용할수 있다

    - 정의할때는 def라는 키워드를 사용
    - def 함수명(파라미터1, 파라미터2, ... 파라미터n)
    
```python
def multiply(a,b):
    s = a*b
    return s
```
   - variable naming과 같이 function의 naming도 특성이나 목적에 맞게 잘 설명할 수 있는 것으로 만들어야 추후에 편하게 쓸수 있다.
   - 위의 예시에서 a, n처럼 함수에서 입력값들을 parameter 또는 argument라고 한다.
   - return으로 반환 값을 지정해줄수 있다

````python
a = [1, 2, 3, 4, 5]
print(len(a))
print(sum(a))

b = -3
print(abs(b))  ## abs -> 절댓값
````

````python
def multiply(a, b):
    s = a+b
    return s  ## return 꼭 해야함, 까먹으면 안됨
````

````python
m = multiply(5, 20)
print(m)
````

````python
def divide(a, b):
    d = a/b
    return d

divide(10, 5)
````

#### **parameter(argument) (인자)**

- 함수에 전달되는 입력 값 (input)
- 입력이 없을수도, 1개 또는 여러개로 입력이 가능할수 있음
- 모든 파이썬 객체 입력 가능 - int, string, float, list등등
- 함수의 파라미터도 가능

````python
def test():
    print('a')
    print('1')
    return 3  ## return 값과 상관없이 무조건 print 안에 있는 값이 출력됨

a = test()
print(a) ## 3으로 출력됨
````

````python
def add(x, y, z):
    add = x + y + z
    return add

a = add('I ', 'love ', 'you')
print(a)
````

````python
## 문자열--> 더할 수는 있지만 뺄 수는 없음
def subtract(x, y):
    sub = x-y
    return sub

a = subtract('apple', 'banana')
print(a)
````

####  Default parameter (기본 인자) 

- 함수의 파라미터는 기본값으로 지정이 가능
- 이때 파라미터를 입력하지 않으면 지정된 기본값이 사용

````python
def add(a, b, c):
    add = a + b + c
    return add

temp = add(10, 20, 30)
print(temp)

def add2(x, y, z = 30):
    add2 = x + y + z
    return add2

temp1 = add2(10, 20, 30)
print(temp1)

temp2 = add2(10, 20)
print(temp2)

temp3 = add2(10, 20, 40) ## z = 40
print(temp3)
````

- Default parameter 사용 시 디폴트 파라미터 뒤에 일반 파라미터가 오면 안됨

````python
def test(a, b, c = 30):
    print(a, b, c)
test(10, 20, 10)

def test(a, b = 10, c = 20):
    print(a, b, c)
    
test(30)
````

````python
def test(a, b = 20, c):
    print(a, b, c)
test(10, 20, 30)
````

````python
## 위에꺼 굳이 하고 싶다면
def test(a, c, b = 20):
    print(a, b, c)
test(10, 20, 30)
````

````python
def test(x, y, z):
    a = x + y + z
    return a

test(10, 20, 3)
# test(x = 10, y = 20, z = 3)
# test(x = 10, z = 3, z = 20)  
````

#### **variable scope ** 
 + 변수가 쓰여지는 범위를 의미
 + 기본적으로 함수내의 변수는 블록이 종료되면 사라짐: **지역변수(local variable)**
 + 처음부터 정해져서 프로그램 종료까지 유지되는 변수: **전역변수(global variable)**
 + 같은 이름의 지역변수와 전역변수가 존재할 경우, 지역변수의 우선순위가 더 높음

````python
num1 = 1
num2 = 5

def test(num1, num2):
    print(num1, num2)
    return num1+num2

test(100, 300)
print(num1, num2)
````

#### **variable length argument (가변길이 인자)**
 - 파라미터 개수가 일정하지 않을때 사용

- ***args**    : 파라미터를 튜플의 형태로 전달

- ****kwargs** : 파리미터를 딕셔너리 형태로 전달

````python
print()
print(1)
print(1, 2)
````

````python
def test(*x):
    print(type(x))

test(10)
test()   ## * 안 붙이면 test() 했을 때 오류 생김
test(1, 2, 3)
````

````python
def test2(*x):
    for item in x:
        print(item)
        
test2(10, 20, 30)
test2(10)
test2() ## 안 넣었을 경우 출력이 안됨
````

````python
def test3(**x):
    print(type(x))
    
test3()
test3(a = 1, b = 2, c = 3, name = 'Mike')
````

###  lambda 함수

 - 단일문으로 표현되는 간단한 함수
 - 주로 반복되지 않는 일회성의 함수를 쓸때 사용

````python
def square(x):
    return x**2

square(5)
````

````python
square2 = lambda x:x**2

square2(5)
````

````python
def add(x, y):
    return x + y

add2 = lambda x, y:x+y
add2(10, 20)
````

````python
(lambda x, y:x+y)(10, 20)
````

#### **filter, map, reduce**
 - lambda가 유용하게 사용되는 3가지 대표적 함수
 - 함수형 프로그래밍의 기본 요소

-  filter : 특정 조건을 만족하는 요소만 남기고 필터링

````python
def even(n):
    return n % 2 == 0

nums = [1, 2, 3, 4, 5, 6, 7]
filter(even, nums)
list(filter(even, nums))
````

 - map: 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환

````python
nums = [1, 2, 3, 4, 5, 6, 7]
list(map(lambda n:n**2, nums))
````

````python
list(map(even, nums))
````

 - reduce : 시퀀스(문자열, 리스트, 튜플)의 원소들을 누적하면서 함수에 적용

````python
from functools import reduce
reduce(lambda x, y: x + y, [0, 1, 2, 3, 4])
````
